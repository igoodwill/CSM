```python
import numpy as np
import matplotlib.pyplot as plt

def normalize(v):
    norm = sum(v)
    if norm == 0:
        return v
    return v / norm

def draw(alpha):
    N = 100

    n = 3

    P = normalize(np.random.rand(n))
    R = normalize(np.random.rand(n))

    Pt = P[:]
    Rt = R[:]

    for k in range(1,N):
        for i in range(0,n):
            P[i]= P[i]*(1-alpha*R[i])
            R[i]= R[i]*(1-alpha*P[i])

        P = normalize(P)
        R = normalize(R)

        Pt = np.c_[Pt,P]
        Rt = np.c_[Rt,R]

    figure, axis = plt.subplots(n, 2)
    for i in range(0,n):
        x = range(0,N)
        y = Pt[i]

        axis[i, 0].plot(x, y)
        axis[i, 0].set_title("Pt")

        y = Rt[i]

        axis[i, 1].plot(x, y)
        axis[i, 1].set_title("Rt")

    plt.show()
```

Alpha = 0.5

```python
draw(0.5)
```

Alpha = -0.5

```python
draw(-0.5)
```
