```python
from math import exp
from random import random
import matplotlib.pyplot as plt

def draw(c):
    N = 100

    x = random()
    xt = [x]

    for k in range(1,N):
        x = c * x * exp(-x)
        xt.append(x)

    plt.plot(range(0,N),xt)
    plt.show()
```

с = 1

```python
draw(1)
```

с = 2

```python
draw(2)
```

с = 5

```python
draw(5)
```

с = 10

```python
draw(10)
```
